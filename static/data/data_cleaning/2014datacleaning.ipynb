```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2014 = "FBI_crime_data_raw/2014.csv"
raw_data2014_df = pd.read_csv(raw_data2014)

raw_data2014_df.columns
```

```python
raw_data2014_df = raw_data2014_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Rape\n (revised\ndefinition)2":"Rape", "Rape\n (legacy\ndefinition)3": "Rape2", "Aggravated\nassault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson4":"Arson"})
raw_data2014_df
```

```python
raw_data2014_df = raw_data2014_df.drop(columns=["Rape2", "Unnamed: 15"])
raw_data2014_df
```

```python
# replace empty strings with NaN
raw_data2014_df['State'] = raw_data2014_df['State'].replace(' ', np.nan).ffill()
raw_data2014_df['City'] = raw_data2014_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2014_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2014_df['City'].unique()
Cities
```

```python
# fix spellings
#raw_data2013_df["State"].replace({"ARKANSAS5": "ARKANSAS", "COLORADO5": "COLORADO", "FLORIDA5": "FLORIDA", "IDAHO5":"IDAHO", "IOWA5": "IOWA", "KENTUCKY5": "KENTUCKY", "MINNESOTA5":"MINNESOTA", "MONTANA5": "MONTANA", "MISSOURI5": "MISSOURI5", "NEW HAMPSHIRE5":"NEW HAMPSHIRE", "NEW YORK6": "NEW YORK","NORTH DAKOTA5": "NORTH DAKOTA", "PENNSYLVANIA5": "PENNSYLVANIA", "RHODE ISLAND5": "RHODE ISLAND", "SOUTH CAROLINA5": "SOUTH CAROLINA", "TENNESSEE5": "TENNESSEE", "UTAH5": "UTAH", "VIRGINIA5": "VIRGINIA", "WISCONSIN5": "WISCONSIN"}, inplace=True)

raw_data2014_df["City"].replace({"BIRMINGHAM5": "BIRMINGHAM", "MOBILE6": "MOBILE", "MONTGOMERY7": "MONTGOMERY", "ANTIOCH8": "ANTIOCH", "CENTENNIAL8": "CENTENNIAL", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "CHICAGO9": "CHICAGO", "ROCKFORD10": "ROCKFORD", "LOUISVILLE METRO": "LOUISVILLE", "DETROIT10":"DETROIT", "LAS VEGAS METROPOLITAN \n  POLICE DEPARTMENT": "LAS VEGAS", "AKRON11": "AKRON", "CINCINNATI11": "CINCINNATI", "CLEVELAND11": "CLEVELAND", "COLUMBUS7, 11": "COLUMBUS", "DAYTON11": "DAYTON", "LUBBOCK8": "LUBBOCK", "MESQUITE8": "MESQUITE", "ROUND ROCK8": "ROUND ROCK", "RICHARDSON8":"RICHARDSON", "SPOKANE8": "SPOKANE"}, inplace=True)

# fill in empty state/city values
raw_data2014_df['State'] = raw_data2014_df['State'].ffill()
raw_data2014_df['City'] = raw_data2014_df['City'].ffill()

raw_data2014_df
```

```python
# drop rows with missing values
data2014_df = raw_data2014_df.dropna()
data2014_df
```

```python
data2014_df.dtypes
```

```python
# Convert data types
data2014_df["Year"] = data2014_df["Year"].astype(int)
data2014_df["Population"] = data2014_df["Population"].str.replace(',', '').astype(int)
data2014_df["Violent Crime"] = data2014_df["Violent Crime"].str.replace(',', '').astype(int)
data2014_df["Murder"] = data2014_df["Murder"].astype(int)
data2014_df["Rape"] = data2014_df["Rape"].str.replace(',', '').astype(int)
data2014_df["Robbery"] = data2014_df["Robbery"].str.replace(',', '').astype(int)
data2014_df["Aggravated Assault"] = data2014_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2014_df["Property Crime"] = data2014_df["Property Crime"].str.replace(',', '').astype(int)
data2014_df["Burglary"] = data2014_df["Burglary"].str.replace(',', '').astype(int)
data2014_df["Larceny Theft"] = data2014_df["Larceny Theft"].str.replace(',', '').astype(int)
data2014_df["Motor Vehicle Theft"] = data2014_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2014_df["Arson"] = data2014_df["Arson"].astype(int)
```

```python
data2014_df
```

```python
# sort by State
data2014_df.sort_values(by=["State"], inplace=True)
data2014_df
```

```python
#export cleaned dataframe
data2014_df.to_csv("FBI_crime_data_cleaned/cleaned_2014.csv", index = False)
data2014_df.to_json("FBI_crime_data_cleaned/cleaned_2014.json")
```
