```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2019 = "FBI_crime_data_raw/2019.csv"
raw_data2019_df = pd.read_csv(raw_data2019)

raw_data2019_df.columns
```

```python
raw_data2019_df = raw_data2019_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent \ncrime":"Violent Crime", "Rape2":"Rape", "Aggravated \nassault":"Aggravated Assault", "Property \ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor \nvehicle \ntheft":"Motor Vehicle Theft", "Arson3":"Arson"})
raw_data2019_df
```

```python
raw_data2019_df = raw_data2019_df.drop(columns=["Unnamed: 14"])
raw_data2019_df
```

```python
# replace empty strings with NaN
raw_data2019_df['State'] = raw_data2019_df['State'].replace(' ', np.nan).ffill()
raw_data2019_df['City'] = raw_data2019_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2019_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2019_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2019_df["State"].replace({"NORTH CAROLINA 5, 6, 8": "NORTH CAROLINA"}, inplace=True)

raw_data2019_df["City"].replace({"HUNTSVILLE4": "HUNTSVILLE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "NAPERVILLE5": "NAPERVILLE", "NEW ORLEANS5": "NEW ORLEANS", "BALTIMORE5":"BALTIMORE", "MINNEAPOLIS6": "MINNEAPOLIS", "ROCHESTER6": "ROCHESTER", "JACKSON7": "JACKSON", "INDEPENDENCE6": "INDEPENDENCE", "LAS VEGAS METROPOLITAN POLICE DEPARTMENT": "LAS VEGAS", "FAYETTEVILLE5":"FAYETTEVILLE", "HIGH POINT5": "HIGH POINT", "CINCINNATI5": "CINCINNATI", "COLUMBUS5": "COLUMBUS", "TOLEDO5, 9": "TOLEDO", "OKLAHOMA CITY6": "OKLAHOMA CITY", "PORTLAND10": "PORTLAND", "METROPOLITAN NASHVILLE POLICE DEPARTMENT": "NASHVILLE", "AUSTIN6": "AUSTIN", "CORPUS CHRISTI6": "CORPUS CHRISTI", "HOUSTON5": "HOUSTON", "LAREDO6": "LAREDO", "PASADENA6": "PASADENA", "RICHARDSON6": "RICHARDSON"}, inplace=True)

# fill in empty state/city values
raw_data2019_df['State'] = raw_data2019_df['State'].ffill()
raw_data2019_df['City'] = raw_data2019_df['City'].ffill()

raw_data2019_df
```

```python
# drop rows with missing values
data2019_df = raw_data2019_df.dropna()
data2019_df
```

```python
data2019_df.dtypes
```

```python
# Convert data types
data2019_df["Year"] = data2019_df["Year"].astype(int)
data2019_df["Population"] = data2019_df["Population"].str.replace(',', '').astype(int)
data2019_df["Violent Crime"] = data2019_df["Violent Crime"].str.replace(',', '').astype(int)
data2019_df["Murder"] = data2019_df["Murder"].astype(int)
data2019_df["Rape"] = data2019_df["Rape"].str.replace(',', '').astype(int)
data2019_df["Robbery"] = data2019_df["Robbery"].str.replace(',', '').astype(int)
data2019_df["Aggravated Assault"] = data2019_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2019_df["Property Crime"] = data2019_df["Property Crime"].str.replace(',', '').astype(int)
data2019_df["Burglary"] = data2019_df["Burglary"].str.replace(',', '').astype(int)
data2019_df["Larceny Theft"] = data2019_df["Larceny Theft"].str.replace(',', '').astype(int)
data2019_df["Motor Vehicle Theft"] = data2019_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2019_df["Arson"] = data2019_df["Arson"].astype(int)
```

```python
data2019_df
```

```python
# sort by State
data2019_df.sort_values(by=["State"], inplace=True)
data2019_df
```

```python
#export cleaned dataframe
data2019_df.to_csv("FBI_crime_data_cleaned/cleaned_2019.csv", index = False)
data2019_df.to_json("FBI_crime_data_cleaned/cleaned_2019.json")
```
