```python
import pandas as pd
import numpy as np
```

```python
# import csv
data11 = "FBI_crime_data_cleaned/cleaned_2011.csv"
data11_df = pd.read_csv(data11)

data12 = "FBI_crime_data_cleaned/cleaned_2012.csv"
data12_df = pd.read_csv(data12)

data13 = "FBI_crime_data_cleaned/cleaned_2013.csv"
data13_df = pd.read_csv(data13)

data14 = "FBI_crime_data_cleaned/cleaned_2014.csv"
data14_df = pd.read_csv(data14)

data15 = "FBI_crime_data_cleaned/cleaned_2015.csv"
data15_df = pd.read_csv(data15)

data16 = "FBI_crime_data_cleaned/cleaned_2016.csv"
data16_df = pd.read_csv(data16)

data17 = "FBI_crime_data_cleaned/cleaned_2017.csv"
data17_df = pd.read_csv(data17)

data18 = "FBI_crime_data_cleaned/cleaned_2018.csv"
data18_df = pd.read_csv(data18)

data19 = "FBI_crime_data_cleaned/cleaned_2019.csv"
data19_df = pd.read_csv(data19)
```

```python
joined_data = [data11_df, data12_df, data13_df, data14_df, data15_df, data16_df, data17_df, data18_df, data19_df]
combined_data = pd.concat(joined_data)
combined_data
```

```python
# sort by State
combined_data.sort_values(by=["State", "City", "Year"], inplace=True)
combined_data
```

```python
combined_data.reset_index(drop=True, inplace=True)
combined_data
```

```python
#export cleaned dataframe
combined_data.to_csv("FBI_crime_data_cleaned/combined_data.csv", index = False)
combined_data.to_json("FBI_crime_data_cleaned/combined_data.json")
```
