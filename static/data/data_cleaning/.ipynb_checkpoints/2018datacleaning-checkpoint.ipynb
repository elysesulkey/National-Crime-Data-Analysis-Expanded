```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2018 = "FBI_crime_data_raw/2018.csv"
raw_data2018_df = pd.read_csv(raw_data2018)

raw_data2018_df.columns
```

```python
raw_data2018_df = raw_data2018_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent \ncrime":"Violent Crime", "Rape2":"Rape", "Aggravated \nassault":"Aggravated Assault", "Property \ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor \nvehicle \ntheft":"Motor Vehicle Theft", "Arson3":"Arson"})
raw_data2018_df
```

```python
raw_data2018_df = raw_data2018_df.drop(columns=["Unnamed: 14"])
raw_data2018_df
```

```python
# replace empty strings with NaN
raw_data2018_df['State'] = raw_data2018_df['State'].replace(' ', np.nan).ffill()
raw_data2018_df['City'] = raw_data2018_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2018_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2018_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2018_df["State"].replace({"NORTH CAROLINA8": "NORTH CAROLINA"}, inplace=True)

raw_data2018_df["City"].replace({"MOBILE4": "MOBILE", "RIALTO5": "RIALTO", "THORNTON6": "THORNTON", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "HONOLULU5": "HONOLULU", "NEW ORLEANS7": "NEW ORLEANS", "LAS VEGAS METROPOLITAN POLICE DEPARTMENT": "LAS VEGAS", "LAS CRUCES5":"LAS CRUCES", "TOLEDO9": "TOLEDO", "DALLAS5": "DALLAS", "GRAND PRAIRIE5": "GRAND PRAIRIE", "HOUSTON5": "HOUSTON", "PROVO6": "PROVO", "RICHMOND7": "RICHMOND"}, inplace=True)

# fill in empty state/city values
raw_data2018_df['State'] = raw_data2018_df['State'].ffill()
raw_data2018_df['City'] = raw_data2018_df['City'].ffill()

raw_data2018_df
```

```python
# drop rows with missing values
data2018_df = raw_data2018_df.dropna()
data2018_df
```

```python
data2018_df.dtypes
```

```python
# Convert data types
data2018_df["Year"] = data2018_df["Year"].astype(int)
data2018_df["Population"] = data2018_df["Population"].str.replace(',', '').astype(int)
data2018_df["Violent Crime"] = data2018_df["Violent Crime"].str.replace(',', '').astype(int)
data2018_df["Murder"] = data2018_df["Murder"].astype(int)
data2018_df["Rape"] = data2018_df["Rape"].str.replace(',', '').astype(int)
data2018_df["Robbery"] = data2018_df["Robbery"].str.replace(',', '').astype(int)
data2018_df["Aggravated Assault"] = data2018_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2018_df["Property Crime"] = data2018_df["Property Crime"].str.replace(',', '').astype(int)
data2018_df["Burglary"] = data2018_df["Burglary"].str.replace(',', '').astype(int)
data2018_df["Larceny Theft"] = data2018_df["Larceny Theft"].str.replace(',', '').astype(int)
data2018_df["Motor Vehicle Theft"] = data2018_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2018_df["Arson"] = data2018_df["Arson"].astype(int)
```

```python
data2018_df
```

```python
# sort by State
data2018_df.sort_values(by=["State"], inplace=True)
data2018_df
```

```python
#export cleaned dataframe
data2018_df.to_csv("FBI_crime_data_cleaned/cleaned_2018.csv", index = False)
```
