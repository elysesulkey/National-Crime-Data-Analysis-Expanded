```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2015 = "FBI_crime_data_raw/2015.csv"
raw_data2015_df = pd.read_csv(raw_data2015)

raw_data2015_df.columns
```

```python
raw_data2015_df = raw_data2015_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Rape\n(revised\ndefinition)2":"Rape", "Rape\n(legacy\ndefinition)3": "Rape2", "Aggravated\nassault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson4":"Arson"})
raw_data2015_df
```

```python
raw_data2015_df = raw_data2015_df.drop(columns=["Rape2", "Unnamed: 15", "Unnamed: 16", "Unnamed: 17"])
raw_data2015_df
```

```python
# replace empty strings with NaN
raw_data2015_df['State'] = raw_data2015_df['State'].replace(' ', np.nan).ffill()
raw_data2015_df['City'] = raw_data2015_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2015_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2015_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2015_df["State"].replace({"OHIO9": "OHIO"}, inplace=True)

raw_data2015_df["City"].replace({"MOBILE5": "MOBILE", "TUCSON6": "TUCSON", "ANAHEIM7": "ANAHEIM", "COSTA MESA7": "COSTA MESA", "DOWNEY7": "DOWNEY", "EL MONTE7": "EL MONTE", "FULLERTON7":"FULLERTON", "GARDEN GROVE7":"GARDEN GROVE",
       "GLENDALE7": "GLENDALE", "INGLEWOOD7": "INGLEWOOD", "IRVINE7":"IRVINE", "ORANGE7":"ORANGE", "PASADENA7":"PASADENA", "POMONA7":"POMONA", "SANTA ANA7":"SANTA ANA", "SANTA ROSA7":"SANTA ROSA", "SIMI VALLEY7":"SIMI VALLEY", "TORRANCE7":"TORRANCE", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "THORNTON8": "THORNTON", "HONOLULU6":"HONOLULU", "LOUISVILLE METRO": "LOUISVILLE", "GRAND RAPIDS10":"GRAND RAPIDS", "TOLEDO8, 11": "TOLEDO", "TYLER6": "TYLER", "PROVO6":"PROVO", "WEST VALLEY6": "WEST VALLEY", "BELLEVUE9":"BELLEVUE"}, inplace=True)

# fill in empty state/city values
raw_data2015_df['State'] = raw_data2015_df['State'].ffill()
raw_data2015_df['City'] = raw_data2015_df['City'].ffill()

raw_data2015_df
```

```python
# drop rows with missing values
data2015_df = raw_data2015_df.dropna()
data2015_df
```

```python
data2015_df.dtypes
```

```python
# Convert data types
data2015_df["Year"] = data2015_df["Year"].astype(int)
data2015_df["Population"] = data2015_df["Population"].str.replace(',', '').astype(int)
data2015_df["Violent Crime"] = data2015_df["Violent Crime"].str.replace(',', '').astype(int)
data2015_df["Murder"] = data2015_df["Murder"].astype(int)
data2015_df["Rape"] = data2015_df["Rape"].str.replace(',', '').astype(int)
data2015_df["Robbery"] = data2015_df["Robbery"].str.replace(',', '').astype(int)
data2015_df["Aggravated Assault"] = data2015_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2015_df["Property Crime"] = data2015_df["Property Crime"].str.replace(',', '').astype(int)
data2015_df["Burglary"] = data2015_df["Burglary"].str.replace(',', '').astype(int)
data2015_df["Larceny Theft"] = data2015_df["Larceny Theft"].str.replace(',', '').astype(int)
data2015_df["Motor Vehicle Theft"] = data2015_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2015_df["Arson"] = data2015_df["Arson"].astype(int)
```

```python
data2015_df
```

```python
# sort by State
data2015_df.sort_values(by=["State"], inplace=True)
data2015_df
```

```python
#export cleaned dataframe
data2015_df.to_csv("FBI_crime_data_cleaned/cleaned_2015.csv", index = False)
```
