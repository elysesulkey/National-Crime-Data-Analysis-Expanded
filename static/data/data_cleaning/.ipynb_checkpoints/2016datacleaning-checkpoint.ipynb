```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2016 = "FBI_crime_data_raw/2016.csv"
raw_data2016_df = pd.read_csv(raw_data2016)

raw_data2016_df.columns
```

```python
raw_data2016_df = raw_data2016_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Rape\n(revised\ndefinition)2":"Rape", "Rape\n(legacy\ndefinition)3": "Rape2", "Aggravated\nassault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson4":"Arson"})
raw_data2016_df
```

```python
raw_data2016_df = raw_data2016_df.drop(columns=["Rape2"])
raw_data2016_df
```

```python
# replace empty strings with NaN
raw_data2016_df['State'] = raw_data2016_df['State'].replace(' ', np.nan).ffill()
raw_data2016_df['City'] = raw_data2016_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2016_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2016_df['City'].unique()
Cities
```

```python
# fix spellings
#raw_data2015_df["State"].replace({"OHIO9": "OHIO"}, inplace=True)

raw_data2016_df["City"].replace({"WATERBURY6": "WATERBURY", "ORLANDO7": "ORLANDO", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "EVANSVILLE8": "EVANSVILLE", "LOUISVILLE METRO9": "LOUISVILLE", "OMAHA10":"OMAHA", "LAS VEGAS METROPOLITAN POLICE DEPARTMENT": "LAS VEGAS", "CARROLLTON11":"CARROLLTON", "EVERETT6": "EVERETT"}, inplace=True)

# fill in empty state/city values
raw_data2016_df['State'] = raw_data2016_df['State'].ffill()
raw_data2016_df['City'] = raw_data2016_df['City'].ffill()

raw_data2016_df
```

```python
# drop rows with missing values
data2016_df = raw_data2016_df.dropna()
data2016_df
```

```python
data2016_df.dtypes
```

```python
# Convert data types
data2016_df["Year"] = data2016_df["Year"].astype(int)
data2016_df["Population"] = data2016_df["Population"].str.replace(',', '').astype(int)
data2016_df["Violent Crime"] = data2016_df["Violent Crime"].str.replace(',', '').astype(int)
data2016_df["Murder"] = data2016_df["Murder"].astype(int)
data2016_df["Rape"] = data2016_df["Rape"].str.replace(',', '').astype(int)
data2016_df["Robbery"] = data2016_df["Robbery"].str.replace(',', '').astype(int)
data2016_df["Aggravated Assault"] = data2016_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2016_df["Property Crime"] = data2016_df["Property Crime"].str.replace(',', '').astype(int)
data2016_df["Burglary"] = data2016_df["Burglary"].str.replace(',', '').astype(int)
data2016_df["Larceny Theft"] = data2016_df["Larceny Theft"].str.replace(',', '').astype(int)
data2016_df["Motor Vehicle Theft"] = data2016_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2016_df["Arson"] = data2016_df["Arson"].astype(int)
```

```python
data2016_df
```

```python
# sort by State
data2016_df.sort_values(by=["State"], inplace=True)
data2016_df
```

```python
#export cleaned dataframe
data2016_df.to_csv("FBI_crime_data_cleaned/cleaned_2016.csv", index = False)
data2016_df.to_json("FBI_crime_data_cleaned/cleaned_2016.json")
```
