```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2017 = "FBI_crime_data_raw/2017.csv"
raw_data2017_df = pd.read_csv(raw_data2017)

raw_data2017_df.columns
```

```python
raw_data2017_df = raw_data2017_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Rape2":"Rape", "Aggravated\nassault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson3":"Arson"})
raw_data2017_df
```

```python
raw_data2017_df = raw_data2017_df.drop(columns=["Unnamed: 14", "Unnamed: 15", "Unnamed: 16", "Unnamed: 17"])
raw_data2017_df
```

```python
# replace empty strings with NaN
raw_data2017_df['State'] = raw_data2017_df['State'].replace(' ', np.nan).ffill()
raw_data2017_df['City'] = raw_data2017_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2017_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2017_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2017_df["State"].replace({"GEORGIA6": "GEORGIA", "NORTH CAROLINA8": "NORTH CAROLINA"}, inplace=True)

raw_data2017_df["City"].replace({"MOBILE4": "MOBILE", "ORLANDO5": "ORLANDO", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "LOUISVILLE METRO7": "LOUISVILLE", "JACKSON8":"JACKSON", "OMAHA6": "OMAHA", "LAS VEGAS METROPOLITAN POLICE DEPARTMENT": "LAS VEGAS", "LAS CRUCES8":"LAS CRUCES", "TOLEDO9, 10": "TOLEDO", "PROVIDENCE11": "PROVIDENCE", "KENOSHA6": "KENOSHA"}, inplace=True)

# fill in empty state/city values
raw_data2017_df['State'] = raw_data2017_df['State'].ffill()
raw_data2017_df['City'] = raw_data2017_df['City'].ffill()

raw_data2017_df
```

```python
# drop rows with missing values
data2017_df = raw_data2017_df.dropna()
data2017_df
```

```python
data2017_df.dtypes
```

```python
# Convert data types
data2017_df["Year"] = data2017_df["Year"].astype(int)
data2017_df["Population"] = data2017_df["Population"].str.replace(',', '').astype(int)
data2017_df["Violent Crime"] = data2017_df["Violent Crime"].str.replace(',', '').astype(int)
data2017_df["Murder"] = data2017_df["Murder"].astype(int)
data2017_df["Rape"] = data2017_df["Rape"].str.replace(',', '').astype(int)
data2017_df["Robbery"] = data2017_df["Robbery"].str.replace(',', '').astype(int)
data2017_df["Aggravated Assault"] = data2017_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2017_df["Property Crime"] = data2017_df["Property Crime"].str.replace(',', '').astype(int)
data2017_df["Burglary"] = data2017_df["Burglary"].str.replace(',', '').astype(int)
data2017_df["Larceny Theft"] = data2017_df["Larceny Theft"].str.replace(',', '').astype(int)
data2017_df["Motor Vehicle Theft"] = data2017_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2017_df["Arson"] = data2017_df["Arson"].astype(int)
```

```python
data2017_df
```

```python
# sort by State
data2017_df.sort_values(by=["State"], inplace=True)
data2017_df
```

```python
#export cleaned dataframe
data2017_df.to_csv("FBI_crime_data_cleaned/cleaned_2017.csv", index = False)
```
