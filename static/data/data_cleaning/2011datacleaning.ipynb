```python
import pandas as pd
import numpy as np
```

```python
# import csvs
raw_data2011 = "FBI_crime_data_raw/2011.csv"
raw_data2011_df = pd.read_csv(raw_data2011)
raw_data2011_df.columns
```

```python
raw_data2011_df = raw_data2011_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Forcible rape":"Rape", "Aggravated assault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson2":"Arson"})
raw_data2011_df
```

```python
raw_data2011_df = raw_data2011_df.drop(columns=["Unnamed: 14", "Unnamed: 15"])
raw_data2011_df
```

```python
# replace empty strings with NaN
raw_data2011_df['State'] = raw_data2011_df['State'].replace(' ', np.nan).ffill()
raw_data2011_df['City'] = raw_data2011_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2011_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2011_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2011_df["State"].replace({"ALABAMA3": "ALABAMA"}, inplace=True)

raw_data2011_df["City"].replace({"BIRMINGHAM4": "BIRMINGHAM", "MOBILE5": "MOBILE", "TUCSON6": "TUSCON", "THORNTON3,6": "THORNTON", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "CHICAGO7": "CHICAGO", "FORT WAYNE3": "FORT WAYNE", "DAVENPORT3": "DAVENPORT", "KANSAS CITY3": "KANSAS CITY3", "LEXINGTON3":"LEXINGTON", "LOUISVILLE METRO": "LOUISVILLE", "STERLING HEIGHTS3": "STERLING HEIGHTS", "ROCHESTER7": "ROCHESTER","INDEPENDENCE3": "INDEPENDENCE", "BILLINGS3":"BILLINGS", "LAS VEGAS METROPOLITAN \n  POLICE DEPARTMENT": "LAS VEGAS", "BUFFALO3": "BUFFALO", "TOLEDO6": "TOLEDO", "EUGENE3": "EUGENE", "GRESHAM3": "GRESHAM", "PORTLAND3": "PORTLAND", "MIDLAND3": "MIDLAND", "GREEN BAY3": "GREEN BAY", "MADISON3": "MADISON"}, inplace=True)

# fill in empty state/city values
raw_data2011_df['State'] = raw_data2011_df['State'].ffill()
raw_data2011_df['City'] = raw_data2011_df['City'].ffill()

raw_data2011_df
```

```python
# drop rows with missing values
data2011_df = raw_data2011_df.dropna()
data2011_df
```

```python
data2011_df.dtypes
```

```python
# Convert data types
data2011_df["Year"] = data2011_df["Year"].astype(int)
data2011_df["Population"] = data2011_df["Population"].str.replace(',', '').astype(int)
data2011_df["Violent Crime"] = data2011_df["Violent Crime"].str.replace(',', '').astype(int)
data2011_df["Murder"] = data2011_df["Murder"].str.replace(',', '').astype(int)
data2011_df["Rape"] = data2011_df["Rape"].str.replace(',', '').astype(int)
data2011_df["Robbery"] = data2011_df["Robbery"].str.replace(',', '').astype(int)
data2011_df["Aggravated Assault"] = data2011_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2011_df["Property Crime"] = data2011_df["Property Crime"].str.replace(',', '').astype(int)
data2011_df["Burglary"] = data2011_df["Burglary"].str.replace(',', '').astype(int)
data2011_df["Larceny Theft"] = data2011_df["Larceny Theft"].str.replace(',', '').astype(int)
data2011_df["Motor Vehicle Theft"] = data2011_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2011_df["Arson"] = data2011_df["Arson"].str.replace(',', '').astype(int)
```

```python
data2011_df
```

```python
# sort by State
data2011_df.sort_values(by=["State"], inplace=True)
data2011_df
```

```python
#export cleaned dataframe
data2011_df.to_csv("FBI_crime_data_cleaned/cleaned_2011.csv", index = False)
data2011_df.to_json("FBI_crime_data_cleaned/cleaned_2011.json")
```
