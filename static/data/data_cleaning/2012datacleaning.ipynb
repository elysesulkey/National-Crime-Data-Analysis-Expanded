```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2012 = "FBI_crime_data_raw/2012.csv"
raw_data2012_df = pd.read_csv(raw_data2012)

raw_data2012_df.columns
```

```python
raw_data2012_df = raw_data2012_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Forcible\nrape":"Rape", "Aggravated\nassault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson2":"Arson"})
raw_data2012_df
```

```python
raw_data2012_df = raw_data2012_df.drop(columns=["Unnamed: 14", "Unnamed: 15"])
raw_data2012_df
```

```python
# replace empty strings with NaN
raw_data2012_df['State'] = raw_data2012_df['State'].replace(' ', np.nan).ffill()
raw_data2012_df['City'] = raw_data2012_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2012_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2012_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2012_df["State"].replace({"ALABAMA3": "ALABAMA", "MONTANA3": "MONTANA"}, inplace=True)

raw_data2012_df["City"].replace({"MOBILE4": "MOBILE", "CHANDLER3": "CHANDLER", "TUCSON5": "TUSCON", "THORNTON3,6": "THORNTON", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "CHICAGO7": "CHICAGO", "DAVENPORT3": "DAVENPORT", "KANSAS CITY3": "KANSAS CITY3", "LEXINGTON3":"LEXINGTON", "LOUISVILLE METRO": "LOUISVILLE", "STERLING HEIGHTS3": "STERLING HEIGHTS", "ROCHESTER7": "ROCHESTER", "JACKSON3": "JACKSON", "LAS VEGAS METROPOLITAN \n  POLICE DEPARTMENT": "LAS VEGAS", "NORTH LAS VEGAS3,6": "NORTH LAS VEGAS", "BUFFALO3": "BUFFALO", "CARY3": "CARY3", "GREENSBORO6": "GREENSBORO", "TOLEDO5": "TOLEDO", "PROVIDENCE6": "PROVIDENCE", "MADISON8": "MADISON"}, inplace=True)

# fill in empty state/city values
raw_data2012_df['State'] = raw_data2012_df['State'].ffill()
raw_data2012_df['City'] = raw_data2012_df['City'].ffill()

raw_data2012_df
```

```python
# drop rows with missing values
data2012_df = raw_data2012_df.dropna()
data2012_df
```

```python
data2012_df.dtypes
```

```python
# Convert data types
data2012_df["Year"] = data2012_df["Year"].astype(int)
data2012_df["Population"] = data2012_df["Population"].str.replace(',', '').astype(int)
data2012_df["Violent Crime"] = data2012_df["Violent Crime"].str.replace(',', '').astype(int)
data2012_df["Murder"] = data2012_df["Murder"].str.replace(',', '').astype(int)
data2012_df["Rape"] = data2012_df["Rape"].str.replace(',', '').astype(int)
data2012_df["Robbery"] = data2012_df["Robbery"].str.replace(',', '').astype(int)
data2012_df["Aggravated Assault"] = data2012_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2012_df["Property Crime"] = data2012_df["Property Crime"].str.replace(',', '').astype(int)
data2012_df["Burglary"] = data2012_df["Burglary"].str.replace(',', '').astype(int)
data2012_df["Larceny Theft"] = data2012_df["Larceny Theft"].str.replace(',', '').astype(int)
data2012_df["Motor Vehicle Theft"] = data2012_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2012_df["Arson"] = data2012_df["Arson"].str.replace(',', '').astype(int)
```

```python
data2012_df
```

```python
# sort by State
data2012_df.sort_values(by=["State"], inplace=True)
data2012_df
```

```python
#export cleaned dataframe
data2012_df.to_csv("FBI_crime_data_cleaned/cleaned_2012.csv", index = False)
data2012_df.to_json("FBI_crime_data_cleaned/cleaned_2012.json")
```
