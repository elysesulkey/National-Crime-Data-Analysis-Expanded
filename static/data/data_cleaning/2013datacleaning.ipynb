```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data2013 = "FBI_crime_data_raw/2013.csv"
raw_data2013_df = pd.read_csv(raw_data2013)

raw_data2013_df.columns
```

```python
raw_data2013_df = raw_data2013_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Forcible\nrape":"Rape", "Aggravated\nassault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson2":"Arson"})
raw_data2013_df
```

```python
raw_data2013_df = raw_data2013_df.drop(columns=["Unnamed: 14", "Unnamed: 15"])
raw_data2013_df
```

```python
# replace empty strings with NaN
raw_data2013_df['State'] = raw_data2013_df['State'].replace(' ', np.nan).ffill()
raw_data2013_df['City'] = raw_data2013_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data2013_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data2013_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data2013_df["State"].replace({"ARKANSAS5": "ARKANSAS", "COLORADO5": "COLORADO", "FLORIDA5": "FLORIDA", "IDAHO5":"IDAHO", "IOWA5": "IOWA", "KENTUCKY5": "KENTUCKY", "MINNESOTA5":"MINNESOTA", "MONTANA5": "MONTANA", "MISSOURI5": "MISSOURI5", "NEW HAMPSHIRE5":"NEW HAMPSHIRE", "NEW YORK6": "NEW YORK",
       "NORTH DAKOTA5": "NORTH DAKOTA", "PENNSYLVANIA5": "PENNSYLVANIA", "RHODE ISLAND5": "RHODE ISLAND", "SOUTH CAROLINA5": "SOUTH CAROLINA", "TENNESSEE5": "TENNESSEE", "UTAH5": "UTAH", "VIRGINIA5": "VIRGINIA", "WISCONSIN5": "WISCONSIN"}, inplace=True)

raw_data2013_df["City"].replace({"BIRMINGHAM3": "BIRMINGHAM", "MOBILE3, 4": "MOBILE", "ANCHORAGE5": "ANCHORAGE", "CHANDLER6": "CHANDLER", "GILBERT5":"GILBERT", "PHOENIX7": "PHOENIX", "TUCSON6, 8": "TUSCON", "THORNTON6": "THORNTON", "NEW HAVEN5": "NEW HAVEN", "STAMFORD5": "STAMFORD", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "CHICAGO9": "CHICAGO", "ROCKFORD5": "ROCKFORD", "FORT WAYNE5": "FORT WAYNE", "INDIANAPOLIS5": "INDIANAPOLIS", "SOUTH BEND5": "SOUTH BEND", "WICHITA5": "WICHITA", "LEXINGTON5":"LEXINGTON", "LOUISVILLE METRO": "LOUISVILLE", "CAMBRIDGE5":"CAMBRIDGE", "LOWELL5":"LOWELL", "SPRINGFIELD5": "SPRINGFIELD", "WORCESTER5": "WORCESTER", "LAS VEGAS METROPOLITAN \n  POLICE DEPARTMENT": "LAS VEGAS", "AKRON5": "AKRON", "CINCINNATI5": "CINCINNATI", "CLEVELAND5": "CLEVELAND", "DAYTON5": "DAYTON", "SALEM5": "SALEM", "ALLENTOWN6": "ALLENTOWN", "TACOMA5": "TACOMA", "VANCOUVER5": "VANCOUVER"}, inplace=True)

# fill in empty state/city values
raw_data2013_df['State'] = raw_data2013_df['State'].ffill()
raw_data2013_df['City'] = raw_data2013_df['City'].ffill()

raw_data2013_df
```

```python
# drop rows with missing values
data2013_df = raw_data2013_df.dropna()
data2013_df
```

```python
data2013_df.dtypes
```

```python
# Convert data types
data2013_df["Year"] = data2013_df["Year"].astype(int)
data2013_df["Population"] = data2013_df["Population"].str.replace(',', '').astype(int)
data2013_df["Violent Crime"] = data2013_df["Violent Crime"].str.replace(',', '').astype(int)
data2013_df["Murder"] = data2013_df["Murder"].str.replace(',', '').astype(int)
data2013_df["Rape"] = data2013_df["Rape"].str.replace(',', '').astype(int)
data2013_df["Robbery"] = data2013_df["Robbery"].str.replace(',', '').astype(int)
data2013_df["Aggravated Assault"] = data2013_df["Aggravated Assault"].str.replace(',', '').astype(int)
data2013_df["Property Crime"] = data2013_df["Property Crime"].str.replace(',', '').astype(int)
data2013_df["Burglary"] = data2013_df["Burglary"].str.replace(',', '').astype(int)
data2013_df["Larceny Theft"] = data2013_df["Larceny Theft"].str.replace(',', '').astype(int)
data2013_df["Motor Vehicle Theft"] = data2013_df["Motor Vehicle Theft"].str.replace(',', '').astype(int)
data2013_df["Arson"] = data2013_df["Arson"].astype(int)
```

```python
data2013_df
```

```python
# sort by State
data2013_df.sort_values(by=["State"], inplace=True)
data2013_df
```

```python
#export cleaned dataframe
data2013_df.to_csv("FBI_crime_data_cleaned/cleaned_2013.csv", index = False)
data2013_df.to_json("FBI_crime_data_cleaned/cleaned_2013.json")
```
