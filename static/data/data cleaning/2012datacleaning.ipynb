```python
import pandas as pd
import numpy as np
```

```python
# import csv
raw_data = "FBI_crime_data_raw/2012.csv"
raw_data_df = pd.read_csv(raw_data)

raw_data_df.columns
```

```python
raw_data_df = raw_data_df.rename(columns={"Unnamed: 2": "Year", "Population1":"Population", "Violent\ncrime":"Violent Crime", "Forcible\nrape":"Rape", "Aggravated assault":"Aggravated Assault", "Property\ncrime":"Property Crime", "Larceny-\ntheft":"Larceny Theft", "Motor\nvehicle\ntheft":"Motor Vehicle Theft", "Arson2":"Arson"})
raw_data_df
```

```python
raw_data_df = raw_data_df.drop(columns=["Unnamed: 14", "Unnamed: 15"])
raw_data_df
```

```python
# replace empty strings with NaN
raw_data_df['State'] = raw_data_df['State'].replace(' ', np.nan).ffill()
raw_data_df['City'] = raw_data_df['City'].replace(' ', np.nan).ffill()
```

```python
# list of values of State column
States = raw_data_df['State'].unique()
States
```

```python
# list of values of city column
Cities = raw_data_df['City'].unique()
Cities
```

```python
# fix spellings
raw_data_df["State"].replace({"ALABAMA3": "ALABAMA", "MONTANA3": "MONTANA"}, inplace=True)

raw_data_df["City"].replace({"MOBILE4": "MOBILE", "CHANDLER3": "CHANDLER", "TUCSON5": "TUSCON", "THORNTON3,6": "THORNTON", "ATHENS-CLARKE COUNTY": "ATHENS-CLARKE", "SAVANNAH-CHATHAM METROPOLITAN": "SAVANNAH-CHATHAM", "CHICAGO7": "CHICAGO", "DAVENPORT3": "DAVENPORT", "KANSAS CITY3": "KANSAS CITY3", "LEXINGTON3":"LEXINGTON", "LOUISVILLE METRO": "LOUISVILLE", "STERLING HEIGHTS3": "STERLING HEIGHTS", "ROCHESTER7": "ROCHESTER", "JACKSON3": "JACKSON", "LAS VEGAS METROPOLITAN \n  POLICE DEPARTMENT": "LAS VEGAS", "NORTH LAS VEGAS3,6": "NORTH LAS VEGAS", "BUFFALO3": "BUFFALO", "CARY3": "CARY3", "GREENSBORO6": "GREENSBORO", "TOLEDO5": "TOLEDO", "PROVIDENCE6": "PROVIDENCE", "MADISON8": "MADISON"}, inplace=True)

# fill in empty state/city values
raw_data_df['State'] = raw_data_df['State'].ffill()
raw_data_df['City'] = raw_data_df['City'].ffill()

raw_data_df
```

```python
# drop rows with missing values
data_df = raw_data_df.dropna()
data_df
```

```python
data_df.dtypes
```

```python
# Convert data types
data_df["Year"] = data_df["Year"].astype(int)
#data_df["Population"] = data_df["Population"].str.replace(',', '').astype(int)
```

```python
# Convert data types
data_df["Year"] = data_df["Year"].astype(int)
data_df["Population"] = data_df["Population"].astype(int)
data_df["Violent Crime"] = data_df["Violent Crime"].astype(int)
data_df["Murder"] = data_df["Murder"].astype(int)
data_df["Rape"] = data_df["Rape"].astype(int)
data_df["Robbery"] = data_df["Robbery"].astype(int)
data_df["Aggravated assault"] = data_df["Aggravated assault"].astype(int)
data_df["Property Crime"] = data_df["Property Crime"].astype(int)
data_df["Burglary"] = data_df["Burglary"].astype(int)
data_df["Larceny Theft"] = data_df["Larceny Theft"].astype(int)
data_df["Motor Vehicle Theft"] = data_df["Motor Vehicle Theft"].astype(int)
data_df["Arson"] = data_df["Arson"].astype(int)
```

```python
data_df
```

```python
# sort by State
cleaned_data_df = data_df.sort_values(by=["State"], inplace=True)
cleaned_data_df
```

```python
#export cleaned dataframe
cleaned_data_df.to_csv("cleaned_2012.csv", index = False)
cleaned_data_df.to_json("cleaned_2012.json")
```
