```python
import pandas as pd 
from data import statesData
import json

crime_df=pd.read_csv("cleaned_crime_data.csv")

crime_df

```

```python
len(list(crime_df["State"].value_counts().index))
```

```python
df_2018=crime_df[crime_df["Year"]==2018]
df_2018.groupby(["State"]).sum().shape
```

```python
df_2019=crime_df[crime_df["Year"]==2019]
df_2019=df_2019.groupby(["State"]).sum()
df_2019["Proportionate Violent Crime %"]=df_2019["Violent Crime"]/df_2019["Population"]*100000

df_2019
```

```python
state_names=[]
for state in list(df_2019.index):
    state_names.append(state.lower().title())
state_names
```

```python
counter=0
states_data=[]
for state in state_names:
    for state_json in statesData["features"]:
        if state_json["properties"]["name"]==state:
            #print(state_json)
            updated_state_json=state_json
            state_df=df_2019[df_2019.index==state.upper()]

            updated_state_json["properties"]["density"]=list(state_df["Proportionate Violent Crime %"]
.values)[0]
            updated_state_json["properties"]["violent_crimes_count"]=list(state_df["Violent Crime"]
.values)[0]
            updated_state_json["properties"]["property_crimes_count"]=list(state_df["Property Crime"]
.values)[0]
            states_data.append(updated_state_json)
            counter+=1
states_data
```

```python
statesData_2019 = {"type": "FeatureCollection", "features": states_data}
```

```python
with open("statesData2019.json", 'w') as outfile:
    json.dump(statesData_2019, outfile)
```
